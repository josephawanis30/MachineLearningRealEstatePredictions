```python
import pandas as pd
```

```python
df = pd.read_csv('final_combined_data.csv')
df.head
```

```python
df.dtypes
```

```python
zip_price = df.groupby('zip').mean()
zip_price = zip_price[['Price']]
zip_price
zip_price2 = zip_price.copy()
```

```python
zip_price.Price.dtype
```

```python
zp_mean = zip_price.mean()
zp_mean
```

```python
zp_med = zip_price.median()
zp_med
```

```python
zip_price['mean_delta'] = zip_price['Price'] - zp_med['Price']
zip_price
```

```python
zip_price['abs_mean_delta'] = zip_price['mean_delta'].abs()
zip_price
```

```python
min_zip = zip_price.abs_mean_delta.min()
```

```python
zip_price.loc[zip_price['abs_mean_delta'] == min_zip]
```

```python
zip_price2['mean_delta'] = zip_price2['Price'] - zp_mean['Price']
zip_price2
```

```python
zip_price2['abs_mean_delta'] = zip_price2['mean_delta'].abs()
zip_price2
```

```python
min_zip2 = zip_price2.abs_mean_delta.min()
zip_price2.loc[zip_price2['abs_mean_delta'] == min_zip2]
```
