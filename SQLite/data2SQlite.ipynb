```python
import pandas as pd
import sqlite3
```

## Calling in Combined Data Set

```python
ml_df = pd.read_csv('../ml_model_data.csv')
ml_df.head()
```

## Calling in Education Data Set

```python
edu_df = pd.read_csv('../education_by_zip.csv')
edu_df.head()
```

## Calling in Crime Data Set

```python
crime_df = pd.read_csv('../Phoenix_crime/Phoenix_Crime_Rate.csv')
crime_df.head()
```

## Calling in the Housing Price Data Set

```python
price_df = pd.read_csv('../Unused_Resources/realty_raw_set.csv', index_col='ID')
price_df.head()
```

```python
# Set up connection
conn=sqlite3.connect('db.sqlite')
```

## Combined Data Set to Housing Data Table

```python
# Convert Pandas DataFrame to SQL Database
ml_df.to_sql('ml_housing_data', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database.
pd.read_sql('select * from ml_housing_data', conn)
```

## Education Data to Education Table

```python
# Convert Pandas DataFrame to SQL Database
edu_df.to_sql('education_scores', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database.
pd.read_sql('select * from education_scores', conn)
```

## Crime Data to Crime Table

```python
# Convert Pandas DataFrame to SQL Database
crime_df.to_sql('crime_rate', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database.
pd.read_sql('select * from crime_rate', conn)
```

## Housing Price Data to Table

```python
# Convert Pandas DataFrame to SQL Database
price_df.to_sql('housing_prices', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database.
pd.read_sql('select * from housing_prices', conn)
```
